```python
#1.3.1
import numpy as np

from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

dot1 = np.array([1.4, 1.4, 1.4])
dot2 = np.array([2.5, 2.5, 2.5])
dot3 = np.array([3.6, 3.6, 3.6])
dot4 = np.array([4.7, 4.7, 4.7])

print(np.linalg.norm(dot1-dot2-dot3-dot4)) # расстояние евклида
print(np.linalg.norm(dot1-dot2-dot3-dot4)**2) # квадрат евклидова расстояния
print(np.linalg.norm(dot1-dot2-dot3-dot4, ord=np.inf)) # расстояние чебышева 
print(np.linalg.norm(dot1-dot2-dot3-dot4, ord=1)) # расстояние хемминга

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

ax.scatter(*dot1)
ax.scatter(*dot2)
ax.scatter(*dot3)
ax.scatter(*dot4)

plt.show()
```

```python
#2.3.1
import numpy as np
import seaborn as sns

from matplotlib import pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

iris = sns.load_dataset('iris')
x_train, x_test, y_train, y_test = train_test_split(iris.iloc[:, :-1],
                                                    iris.iloc[:, -1],
                                                    test_size = 0.15) 

model = KNeighborsClassifier(n_neighbors=10) # 1/5/10
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

plt.figure(figsize=(10, 7))
sns.scatterplot(x="petal_width",y = "petal_length",data=iris,hue='species',s=70)

plt.xlabel("petal length, sm")
plt.ylabel("petal width, sm")
plt.legend(loc=2)
plt.grid()

for i in range(len(y_test)):
    if np.array(y_test)[i] != y_pred[i]:
        plt.scatter(x_test.iloc[i, 3], x_test.iloc[i, 2], color="red", s=150)
        
print(y_pred)
print(f'accuracy: {accuracy_score(y_test, y_pred) :.3}')

plt.show()
```
