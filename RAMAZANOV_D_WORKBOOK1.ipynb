```python
#4.3.1
import numpy as np
from matplotlib import pyplot as plt

array = np.random.sample(10)
print(array)
print(f"average: {np.mean(array)}")
print(f"meadiana: {np.median(array)}")

#Эти два значения близки друг к другу, так как минимальное и максимальное значение - 0 и 1 - очень близки.
#Поэтому их значение в медиане и среднем близки, так как выборка сильно ограничена.

# точечный график 
plt.scatter(array, array)
# настройка осей
plt.gca().set(xlim=(0, 1), ylim=(0, 1))
plt.show()
```

```python
#4.3.2
import math
from matplotlib import pyplot as plt

# подстановка х в данную функцию
def calc_func(x):
    part1 = math.sqrt(1 + math.exp(math.sqrt(x)) + math.cos(x**2))
    part2 = abs(1 - math.sin(x)**3)
    part3 = math.log(2*x)
    return (part1 / part2) + part3

d = list(range(1, 11))
calc_d = [calc_func(x) for x in d]

# создаем нескоько пространств для графиков subplot(nrows, ncols, index)
plt.subplot(1, 2, 1)
plt.plot(d, calc_d)
plt.subplot(1, 2, 2)
plt.scatter(d[:5], calc_d[:5])
plt.show()
```

```python
#4.3.4
from matplotlib import pyplot as plt
import pandas as pd

plt.rcParams["figure.figsize"] = (15, 10)

APPLE = "https://raw.githubusercontent.com/skirt-owner/workbook1_data/main/AAPL.csv"
GOOGLE = "https://raw.githubusercontent.com/skirt-owner/workbook1_data/main/GOOG.csv"
MICROSOFT = "https://raw.githubusercontent.com/skirt-owner/workbook1_data/main/MSFT.csv"
columns = ["Date","High","Low"]
apple_csv = pd.read_csv(APPLE, usecols = columns)
google_csv = pd.read_csv(GOOGLE, usecols = columns)
microsoft_csv = pd.read_csv(MICROSOFT, usecols = columns)

# для удобства редактирования графиков - получим редатируемые области
fig, axs = plt.subplots(3, 1)
# добавляем разрыв между графиками
fig.tight_layout(pad=5)

axs[0].plot(apple_csv.Date, apple_csv.High, label="high")
axs[0].plot(apple_csv.Low, label="low")
axs[0].legend()
axs[0].set_title("APPLE")

axs[1].plot(google_csv.Date, google_csv.High, label="high")
axs[1].plot(google_csv.Low, label="low")
axs[1].legend()
axs[1].set_title("GOOGLE")

axs[2].plot(microsoft_csv.Date, microsoft_csv.High, label="high")
axs[2].plot(microsoft_csv.Low, label="low")
axs[2].legend()
axs[2].set_title("MICROSOFT")

plt.show()
```

```python
#4.3.5
import math

x = float(input("Enter x -> "))
y = float(input("Enter y -> "))

summary = lambda x, y: x+y
subtraction = lambda x, y: x-y
divide = lambda x, y: x/y
multiply = lambda x, y: x*y
exponent = lambda x, y: math.exp(x+y)
sinus = lambda x, y: math.sin(x+y)
cosinus = lambda x, y: math.cos(x+y)
power = lambda x, y: x**y

oper_func = {"+": summary, "-": subtraction, "*": multiply, "/": divide, "exp": exponent, "sin": sinus, "cos": cosinus, "pow": power}

oper = input("'+', '-', '*', '/', 'exp', 'sin', 'cos', 'pow'")

result = oper_func[oper](x, y)

print(f"{x} {oper} {y} = {result}")
    
```
